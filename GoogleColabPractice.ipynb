```python
#EDITS MADE HERE ()
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure with specific size
plt.figure(figsize=(6, 2))

# First subplot
ax1 = plt.subplot(1, 2, 1, projection='3d')
x, y = np.meshgrid(np.arange(0, 3.1, 0.1), np.arange(0, 3.1, 0.1))
r = np.sin(x - 0.5)**2 + np.cos(y - 0.5)**2
ax1.plot_surface(x, y, r, cmap='viridis')

# Second subplot
ax2 = plt.subplot(1, 2, 2, projection='3d')
X = np.linspace(-8, 8, 100)
Y = np.linspace(-8, 8, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax2.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Fills and Alpha Example")
plt.show()
```

```python
#EDIT MADE HERE
import numpy as np
import matplotlib.pyplot as plt

# Generate data
y1 = 10 + np.random.randn(10)
y2 = 10 + np.random.randn(10)
x = np.arange(len(y1))

# Create plots
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.plot(x, y1)
plt.fill_between(x, y1, 7, where=(y1 > 5), facecolor='r', alpha=0.9)

plt.subplot(222)
plt.plot(x, y2)
plt.fill_between(x, y2, 7, where=(y2 > 5), facecolor='r', alpha=0.2)

plt.tight_layout()
plt.show()
```
